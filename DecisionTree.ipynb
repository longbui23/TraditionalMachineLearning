# Decision Trees 

```python
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
my_data = pd.read_csv("drug200.csv", delimiter=",")
my_data[0:5]
```

```python
my_data.shape
```

<div href="pre-processing">
    <h2>Data preprocessing</h2>
</div>


Using <b>my_data</b> as csv data read by pandas Drug.csv, declare the following variables: <br>

<ul>
     <li> <b> X </b> is <b> Feature matrix </b> (data of my_data) </li>
     <li> <b> y </b> is <b> the response vector (destination) </b> </li>
</ul>

Remove the column containing the destination name because it does not contain numeric values.

```python
X = my_data[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']].values
X[0:5]
```

```python
from sklearn import preprocessing
le_sex = preprocessing.LabelEncoder()
le_sex.fit(['F','M'])
X[:,1] = le_sex.transform(X[:,1]) 


le_BP = preprocessing.LabelEncoder()
le_BP.fit([ 'LOW', 'NORMAL', 'HIGH'])
X[:,2] = le_BP.transform(X[:,2])


le_Chol = preprocessing.LabelEncoder()
le_Chol.fit([ 'NORMAL', 'HIGH'])
X[:,3] = le_Chol.transform(X[:,3]) 

X[0:5]
```

```python
y = my_data["Drug"]
y[0:5]
```

<hr>

<div id="setting_up_tree">
    <h2>Setting up Tree</h2>
   

```python
from sklearn.model_selection import train_test_split
```

```python
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
print('Shape of X training set {}'.format(X_trainset.shape),'&',' Size of Y training set {}'.format(y_trainset.shape))
```

print('Shape of X training set {}'.format(X_testset.shape),'&',' Size of Y training set {}'.format(y_testset.shape))

Data Modelling

```python
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
drugTree
```

```python
drugTree.fit(X_trainset,y_trainset)
```

# Predictions

```python
predTree = drugTree.predict(X_testset)
```

```python
print (predTree [0:5])
print (y_testset [0:5])
```

<hr>

<div id="evaluation">
    <h2>Evaluation</h2>

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, predTree))
```

**Accuracy classifier** calculates subset accuracy: the set of predicted labels for a sample must exactly match the set of corresponding labels in y_true.

In a multi-label classifier, the function will return the precision of the subset. If the entire set of labels predicted for a sample perfectly matches the set of real labels, then the subset accuracy is 1.0; otherwise it is 0.0.
