```python
### we will notice that we have 11 values that are empty but encoded as ' '
df.TotalCharges.value_counts()
```

```python
df['TotalCharges'][df.TotalCharges== ' ']='-1'
df['TotalCharges']= pd.to_numeric(df['TotalCharges'])
df['TotalCharges'][df.TotalCharges== -1]=df['TotalCharges'].median()
```

```python
###mapping
mapping= {'Yes':1,'No':0,'No phone service':0,'No internet service':0}
gender_mapping={'Female':1,'Male':0}
```

```python
df.head()
```

```python
df.Partner.value_counts()
```

```python
df.head()
```

```python
df['gender']=df['gender'].map(gender_mapping)
df['Partner']=df['Partner'].map(mapping)
df['Dependents']=df['Dependents'].map(mapping)
df['PhoneService']=df['PhoneService'].map(mapping)
df['MultipleLines']=df['MultipleLines'].map(mapping)
df['OnlineSecurity']=df['OnlineSecurity'].map(mapping)
df['OnlineBackup']=df['OnlineBackup'].map(mapping)
df['DeviceProtection']=df['DeviceProtection'].map(mapping)
df['TechSupport']=df['TechSupport'].map(mapping)
df['StreamingTV']=df['StreamingTV'].map(mapping)
df['StreamingMovies']=df['StreamingMovies'].map(mapping)
df['PaperlessBilling']=df['PaperlessBilling'].map(mapping)
df['Churn']=df['Churn'].map(mapping)
```

```python
df.loc[0,:]
```

```python
df['InternetService']=df['InternetService'].astype('category')
df['Contract']=df['Contract'].astype('category')
df['PaymentMethod']=df['PaymentMethod'].astype('category')
```

```python
df_dumm= pd.get_dummies(df.loc[:,'gender':])
df_scaled=(df_dumm-df_dumm.min())/(df_dumm.max()-df_dumm.min())
```

```python
df_dumm.columns
```

```python
###shuffle the data
df_shuffled= df_scaled.sample(frac=1)
```

```python
df_shuffled.info()
```

```python
x_train=df_shuffled.loc[:6000,:].drop('Churn',axis=1)
x_test=df_shuffled.loc[6001:,:].drop('Churn',axis=1)
y_train=df_shuffled.loc[:6000,'Churn']
y_test=df_shuffled.loc[6001:,'Churn']
```

```python
import statsmodels.api as sm

```

```python
model_logit= sm.Logit(y_train,sm.add_constant(x_train)).fit()
```

```python
print(model_logit.summary())
```

```python
yhat= model_logit.predict(sm.add_constant(x_test))
```

```python
yhat
```

```python
y_all= list(map(round,yhat))
```

```python
from sklearn.metrics import (confusion_matrix,accuracy_score)
```

```python
tn, fp, fn, tp = confusion_matrix(y_test,y_all).ravel()
```

```python
(tn, fp, fn, tp)
```

```python
accuracy_score(y_test,y_all)
```

```python
(y_all == y_test).mean()
```

```python
## precision and recall

precsion= tp/(tp+fp)
precsion


```

```python
recall= tp/(tp+fn)
recall
```
